```python
#import all the required libraries
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)
```

```python
url = "https://flight.yatra.com/air-search-ui/dom2/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=DEL&originCountry=IN&destination=BOM&destinationCountry=IN&flight_depart_date=23%2F10%2F2021&ADT=1&CHD=0&INF=0&class=Economy&source=fresco-home&unqvaldesktop=824008927864"
driver.get(url)
time.sleep(30)
```

```python
airline =[]
source =[]
destination = []
departure_time=[]
arrival_time = []
duration = []
total_stops =[]
price =[]
```

```python
airlines=driver.find_elements_by_xpath("//div[@class='fs-13 airline-name no-pad col-8']/span")
for i in airlines:
    if i.text is None :
        airline.append("--") 
    else:
        airline.append(i.text)
sources=driver.find_elements_by_xpath("//div[@class='i-b col-4 no-wrap text-right dtime col-3']")
for i in sources:
    if i.text is None :
        source.append("--") 
    else:
        source.append(i.text)
departures=driver.find_elements_by_xpath("//div[@class='grid']/div[3]")
for i in departures:
    if i.text is None :
        destination.append("--") 
    else:
        destination.append(i.text)
total_time =driver.find_elements_by_xpath("//div[@class='stop-det bdr-btm pb-5 v-aligm-t text-left']/div/p")
for i in total_time:
    if i.text is None :
        duration.append("--") 
    else:
        duration.append(i.text)
stops =driver.find_elements_by_xpath("//div[@class=' font-lightgrey fs-10 tipsy i-b fs-10']")
for i in stops:
    if i.text is None :
        total_stops.append("--") 
    else:
        total_stops.append(i.text)
fare =driver.find_elements_by_xpath("//div[@class='i-b tipsy fare-summary-tooltip fs-18']")
for i in fare:
    if i.text is None :
        price.append("--") 
    else:
        price.append(i.text)
```

```python
df1=pd.DataFrame({'airline':airline,
                 'source':source,
                'destination':destination,
                'duration':duration,
                'total_stops':total_stops,
                'price':price})
```

```python
df1['date_of_journey'] = '23-10-2021'
```
